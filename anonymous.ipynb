```python
#find the square of a number in a list using lambda and map
lst = [1,2,3,4,5]
#map(fun,iterable)
res_list=list(map(lambda x: x**2,lst))
print(res_list)
```

```python
#even numbers in single line statement
"Even" if 10%2==0 else "Odd"
```

```python
#even numbers using lambda and filter
lst = [1,2,3,4,5]
res_list=list(filter(lambda x: x%2 == 0, lst))
print(res_list)
```

```python
#add 1 to number if number is even, else add 2 if the number is odd
lst = [1,2,3,4,5]
res_list=list(map(lambda x: x+1 if x%2 == 0 else x+2, lst))
print(res_list)
```

```python
#nested if-else
num=-19
("Negative even" if num < 0 else "Positive even") if num % 2 == 0 \
else ("Positive odd" if num > 0 else "Negative odd")
```

```python
#sorted with lambda
string = "Hello world12@#"
dict={}
for char in string.lower():
    if char.isalpha():
        if char in dict:
            dict[char] += 1
        else:
            dict[char] = 1
print(dict)
```
