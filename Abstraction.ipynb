```python
from abc import ABC, abstractmethod
class Car(ABC):
    #abstract methods
    @abstractmethod
    def engine(self): # Method declaration and without implementation
        pass
    @abstractmethod
    def tyres(self):
        pass
    @abstractmethod
    def brakes(self):
        pass

# child class
class Honda(Car):
    def engine(self):
        return "This is honda engine"
    def tyres(self):
        return "Honda car has 4-bit tyres"
    def brakes(self):
        return "Honda car applies brakes automatically"
    def ac(self):
        return "Honda car has central AC"
h = Honda()
print(h.tyres())
print(h.ac())
```
