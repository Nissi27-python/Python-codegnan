```python
#syntax: [expression for iterator in iterable if condition]
```

```python
#list of n numbers
n = int(input())
lst=[]
for i in range(1, n+1):
    lst.append(i)
print(lst)
```

```python
#list of n numbers using list comprehension
n = int(input())
res=[x for x in range(1,n+1)]
print(res)
```

```python
#list of even numbers using list comprehension
n=int(input())
res=[x for x in range(1,n+1) if x%2 == 0]
print(res)
```

```python
#list of n even numbers using list comprehension
n=int(input())
res=[x*2 for x in range(n)]
print(res)
```

```python
#list of odd numbers using list comprehension
n=int(input())
res=[x for x in range(1,n+1) if x%2 != 0]
print(res)
```

```python
#list of n odd numbers using list comprehension
n=int(input())
res=[x*2+1 for x in range(n)]
print(res)
```

```python
# even numbers between range
n1,n2=map(int,input().split())
res=[x for x in range(n1,n2) if x% 2 == 0]
print(res)
          
```

- syntax for if else in list comprehension:
- syntax: [true_exp if condition else false_exp for iterator in iterable]

```python
#list of numbers convert into 1's and 0's.
#even number - 0
#odd number - 1
lst=list(map(int,input().split()))
res = [0 if num % 2 == 0 else 1 for num in lst]
print(res)
```

```python
#even number - leftshift by 1
#odd number - right shift by 1
lst=list(map(int,input().split()))
res = [num<<1 if num % 2 == 0 else num>>1 for num in lst]
print(res)
```

```python
# two dice combination
res = [(x,y) for x in range(1,7) for y in range(1,7)]
print(res)
```

```python
#sum of two dice faces is even
res = [(x,y) for x in range(1,7) for y in range(1,7) if (x+y)%2==0]
print(res)
```

```python
#two dice faces are same
res = [(x,y) for x in range(1,7) for y in range(1,7) if (x==y)]
print(res)
```

```python
#any dice contains 4
res = [(x,y) for x in range(1,7) for y in range(1,7) if x==4 or y==4]
print(res)
```

```python
#combinations of even = 0, combinations of odd = 1
#even sum - 0
#odd sum -1
res = [0 if (x+y)%2==0 else 1 for x in range(1,7) for y in range(1,7)]
print(res)
```

```python
#count of all combinations and which is k != (sum of three dices)
k = int(input())
res = [(x,y,z) for x in range(1,7) for y in range(1,7) for z in range(1,7) if (x+y+z) != k]
print(len(res))
```

```python
#count of all combinations and which is k != (sum of three dices)
k = int(input())
res = sum([1 for x in range(1,7) for y in range(1,7) for z in range(1,7) if (x+y+z) != k])
print(res)
```

```python
#dict comprehension
res ={x:x**2 for x in range(1,7)}
print(type(res))
print(res)
```

```python
#set comprehension
res ={x**2 for x in range(1,7)}
print(type(res))
print(res)
```

```python
#tuple comprehension
res =tuple(x**2 for x in range(1,7))
print(type(res))
print(res)
```
