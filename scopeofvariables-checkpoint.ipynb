```python
x=10
def outer():
    y=20 #local
    print(y)
    def inner():
        z=30 #non local
        print(z)
    inner()
outer()
print(x)
```

```python
# using the global scope variable in local scope and non local scope
x=10#global variable
def outer():
    y=20 #local scope variable
    a=y+x
    def inner():
        z=30 #non local
        b=x+y+z #b is non local scope variable
        print("non local scope:",b)
    inner()
    print("local scope:",a)
outer()
print("Global scope:",x)
```

```python
#defining scope of variables
x=10 # global variable
def outer():
    x=20 #local
    x=x+1
    def inner():
        x=30 #non local
        x=x+2
        print("Non local scope:",x)
    inner()
    print("local scope:",x)
outer()
print("Global scope:",x)
```

```python
# global keyword
x=10
def outer():
    global x # we can update the variable in local
    x=x+1
    def inner():
        #x=30 #non local
        y=x+2
        print("Non local scope:",y)
    inner()
    print("local scope:",x)
outer()
print("Global scope:",x)
```
