```python
#1.without parameter and without return value
#- The function definition doesn't take parameters and doesn't return any value or expression
#- The function definition execution is terminate when return statement is executed
#- The function definition returns some expression otherwise None
#function definition
def sum_of_two_numbers():
    x = 10
    y = 20
    c = x + y
    print("This is from function definition:",c)
    return #function termination point
    print(50) # This statement never execute
#function call
res = sum_of_two_numbers()
print("This is from function call:",res)
```

```python
#patterns using functions
def f_pattern(n):
 for i in range(5):
    for j in range(6):
       if j==0 or i==0 or i==n//2:
            print("*",end = " ")
    print()
f_pattern(5)

     
```

```python
#E pattern using functions
def e_pattern(n):
 for i in range(5):
    for j in range(6):
       if j==0 or i==0 or i==n//2 or i==n-1:
            print("*",end = " ")
    print()
e_pattern(5)

     
```

```python
def c_pattern(n):
    for i in range(n):
        for j in range(n):
            if (i == 0 and j >= 1 and j <=4 ) or (i == n - 1 and 1 <= j <= 4 ) or (j==0 and 1 <= i <= 3):
                print("*", end=" ")
            else:
                print(" ", end=" ")
        print()

# give n value before calling
c_pattern(5)
```

```python
#2.without parameters and with return
def sum_of_two_numbers():
    x = 10
    y = 20
    c = x + y
    return c#function termination point
#function call
res = sum_of_two_numbers()
print("This is from function call:",res)
```

```python
#3.with parameters and without return
def sum_of_two_numbers(num1, num2):
    c = num1 + num2
    print("This is from function definition:",c)
    return #function termination point
#function call
res = sum_of_two_numbers(10, 20)
print("This is from function call:",res)
```

```python
#3.with parameters and with return
def sum_of_two_numbers(num1, num2):
    c = num1 + num2
    print("This is from function definition:",c)
    return c#function termination point
#function call
res = sum_of_two_numbers(10, 20)
print("This is from function call:",res)
```

```python
def even_number(num):
    if num % 2 == 0:
        return True
    else:
        return False
even_number(18)
```
