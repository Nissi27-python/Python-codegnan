**Inheritance**: A class can inherit the properties of another class(parent)
- Here child class object can access the child class properties as well as parent class properties
- Parent class object can access parent class properties only. It can't access child class properties

**Types of Inheritance**
1. Single Inheritance
2. Multi level Inheritance
3. Multiple Inheritance
4. Heirarchical Inheritance
5. Hybrid Inheritance


**Single Inheritance**: One child class can inherit the properties from one parent class
- In single inheritance it have only one parent class and one child class
*Syntax*:
```class parent():```<br>
```   #properties```<br>
```class Child(Parent):```<br>
```   #properties```

````python
# constructor in parent class and not in child class
#here parent class constructor can also act as child's constructor
class Animal:
    def __init__(self, no_legs:int):
        self.no_legs = no_legs
    def legs(self):
        return f"This Animal has {self.no_legs} legs"
#child class implementation
class Dog(Animal):
    def sound(self):# instance method
        return "Dog can make sound as Boww Boww"
a = Animal(4)
d = Dog(4)
print(d.sound())
print(d.legs())
print(a.legs())
print(d.no_legs)
````

````python
# constructor in both parent class and not in child class

class Animal:
    def __init__(self, no_legs:int):
        self.no_legs = no_legs
    def legs(self):
        return f"This Animal has legs"
#child class implementation
class Dog(Animal):
    def __init__(self, sound:str):
        self.sound = sound
    def sounds(self):# instance method
        return f"Dog can make sound like {self.sound}"
a = Animal(4)
d = Dog("Boww Boww")
print(d.sounds())
print(d.legs())
#print(a.legs())
#print(d.no_legs)
````

````python
#inheriting the parent class constructor
# constructor in both parent class and not in child class
class Animal:
    def __init__(self, no_legs:int):
        self.no_legs = no_legs
    def legs(self):
        return f"This Animal has {self.no_legs} legs"
#child class implementation
class Dog(Animal):
    def __init__(self, sound:str, no_legs:int):
        super().__init__(no_legs)
        self.sound = sound
    def sounds(self):# instance method
        return f"Dog can make sound like {self.sound}"

d = Dog("Boww Boww", 4)
print(d.sounds())
print(d.legs())
#print(a.legs())
#print(d.no_legs)
````
