```python
#pass by reference
def fun(l, val):
    l.append(val)
    return l
lst = [1,2,3]
val = 10
print(fun(lst, val))
print(lst)
```

```python
#pass by value
def fun(x):
   x=x+1
   return x
val = 10
print(fun(val))
print(val)
```

```python
#factorial
def fact(n):
    if n == 0:
        return 1
    else:
        return n*fact(n-1)
print(fact(5))
```

```python
#number sequence
def num(n):
    if n == 1:
        return
    print(n)
    num(n-1)
num(5)
```

```python
#number sequence
def num(n):
    if n == 0:
        return
    num(n-1)
    print(n)
num(5)
```

```python
#even numbers 
def even_num(n):
    if n == 0:
        return
    even_num(n-1)
    if n%2==0:
        print(n,end=" ")
even_num(10)
```

```python
#even numbers between n1 and n2
def even_num(n1,n2):
    if n1 > n2:
        return
    if n1 %2 == 0:
        print(n1,end=" ")
    even_num(n1 + 1, n2)
even_num(10,20)
```

```python
#sum of n natural numbers
def natural_num(n):
   if n==0:
       return 0
   else:
       return n + natural_num(n-1)
print(natural_num(20))
```

```python
#odd numbers
def odd_num(n):
    if n == 0:
        return
    odd_num(n-1)
    if n%2!=0:
        print(n,end=" ")
odd_num(10)
```

```python
#sequence functions
#reversed - it reverse the sequence
s="Codegnan"
print(list(reversed(s)))
print("".join(list(reversed(s))))
```

```python
#sorted()-it sorts the sequence
lst=["Apple","Ball","Cat","Preksha"]
print(sorted(lst))
print(sorted(lst,reverse=True))
print(sorted(lst, key = len))
print(sorted(lst, key = len, reverse=True))
```

```python
#enumerate()-it returns a sequence as list of tuple of index and value pairs
lst=["Apple","Ball","Cat","Preksha"]
print(enumerate(lst))
print(list(enumerate(lst)))
```

```python
#any()-This function returns true if any one element in sequence is non zero or non empty sequence
lst=[1,2,'Hi',1,0,0,0]
print(any(lst))
print(any([0,0,0,0,0]))
print(any([2,0,0,0,0,0]))
print(any(['nissi',0]))
```

```python
#all()- this function returns True if all elements in sequence is non zero or non empty sequence
lst=[1,2,'Hi',1,0,0,0]
print(all(lst))
print(all([0,0,0,0,0]))
print(all([2,0,0,0,0,0]))
print(all(['nissi',""]))
print(all([1,2,3,4,5]))
```

```python
#map(fun, seq)-it applies the function on each element in sequence
def add_one(a):
    return a+1
lst=[10,30,40,50]
print(list(map(add_one, lst)))
```

```python
#filter(fun, seq)-it filters the sequence of elements based on function
def even_numbers(x):
    if x%2==0:
        return x
lst=[10,201,301,40]
print(list(map(even_numbers, lst)))
print(list(filter(even_numbers, lst)))
```

```python
#zip(sequences)- it returns list of tuple of all sequence elements
l1=[1,2,3,4]
l2=['a','b','c','d']
print(list(zip(l1,l2)))
```
