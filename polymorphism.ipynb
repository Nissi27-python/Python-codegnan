```python
class Parent:
    def __init__(self, name):
        self.name = name
    def show(self):
        return f"My name is {self.name}"
class Child(Parent):
    def __init__(self, name, job):
        super().__init__(name)
        self.job = job
    def show(self):
        return f"I am working as a {self.job}"
c = Child("Nissi","Trainee")
print(c.show())
   
```

```python
#method overriding achieve through the 'super'
# parent class method executing through the 'super'
class Parent:
    def __init__(self, name):
        self.name = name
    def show(self):
        return f"My name is {self.name}"
class Child(Parent):
    def __init__(self, name, job):
        super().__init__(name)
        self.job = job
    def show(self):
        print(super().show())
        return f"I am working as a {self.job}"
c = Child("Nissi","Trainee")
print(c.show())
```

```python
class Math:
    def add(self, a, b):
        return a+b
    def add(self, a, b, c):
        return a+b+c
    def add(self, a,b,c,d):
        return a+b+c+d
m = Math()
print(m.add(10,20,30,40))
print(m.add(2,5,6))
print(m.add(3,5))

```

```python
# method overloading can achieve through default arguments or variable length of 
# - positional arguments or variable length of keyword arguments
## Method overloading
class Math:
    def add(self, a, b, c=0, d=0):
        return a+b+c+d
m = Math()
print(m.add(10,20,30,40))
print(m.add(2,4,6))
print(m.add(3,5))
```

**Duck Typing**

```python
class Dog:
    def sound(self):
        return "Boww Boww"
class Cat:
    def sound(self):
        return "Meow Meow"
def animal_sound(animal):
    print(animal.sound())
#calling function
animal_sound(Dog())
animal_sound(Cat())
```

**Operator Overloading**

```python
class Num:
    def __init__(self, n):
        self.n = n

n1 = Num(10)
n2 = Num(5)
print(n1.n + n2.n)
print(n1 + n2)
        
```

```python
class Num:
    def __init__(self, n):
        self.n = n
    # magic methods
    def __add__(self, other):
        return Num(self.n + other.n)
    def __repr__(self):
        return f"The num value is {self.n}"
   
    
n1 = Num(10)
n2 = Num(5)
n3 = Num(6)
n4 = n1 + n2
n5 = n4 + n3
print(n5)
```

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def __add__(self, other):
        return Point(self.x + other.x, self.y + other.y)
    def __str__(self):
        return f"The point value is {self.x, self.y}"

p1 = Point(5,4)
p2 = Point(3,2)
p3 = p1 + p2
print(p3)
```
