```python
#user login
users = {1234:456, 1235:567}
def login(username:int, password:int):
    if username in users:
        if password == users[username]:
            print('Login successful')
        else:
            print('incorrect details')
    else:
        print('User not found')
username=int(input())
password=int(input())
login(username = username, password = password)
```

```python
# users = {username: [password, attempts_left, blocked_flag]}
users = {1234: [456, 3, False]}
```

```python

def login(username: int, password: int):
    # Check if username exists
    if username in users:
        # Check if user is blocked
        if not users[username][2]:
            # Check password
            if password == users[username][0]:
                print("Login successful")
                # Reset attempts after successful login
                users[username][1] = 3
            else:
                print("Incorrect details")
                users[username][1] -= 1
                # If attempts become 0, block the user
                if users[username][1] == 0:
                    users[username][2] = True
                    print("User blocked after 3 incorrect attempts")
                else:
                    print(f"Remaining attempts: {users[username][1]}")
        else:
            print("User blocked for 24 hours")
    else:
        print("User not found")
username=int(input())
password=int(input())
login(username,password)
```
