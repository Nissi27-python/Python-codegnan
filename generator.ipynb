```python
def gen():
    x = 5
    while x > 0:
        yield x
        x = x - 1
res = gen()
print(list(res))
print(next(res))
print(next(res))
print(next(res))
```

```python
#to overcome the stopIteration error: 
def gen():
    x = 5
    while x > 0:
        yield x
        x = x-1
res = gen()
for i in res:
    print(i)
```

```python
#checking memory space how much function takes to generate 100000 numbers
def fun():
    res = []
    for i in range(1, 100000):
        res.append(i)
    return res
res = fun()
print(sys.getsizeof(res))
```

```python
#checking memory space how much function takes to generate 100000 numbers
def gen():
    res = []
    for i in range(1, 100000):
        yield i
res = gen()
print(sys.getsizeof(res))
```

```python
lst = [i for i in range(1, 100000)]
print(type(lst))
print(sys.getsizeof(lst))
#shortcut way to create generator
res = (i for i in range(1, 100000))
print(type(res))
print(sys.getsizeof(res))
```
