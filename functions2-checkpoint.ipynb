```python
def is_palindrome(num):
    original = num
    res=0
    while num > 0:
      rem = num % 10
      res = res * 10 + rem
      num = num // 10
    return res == original
print(is_palindrome(121)) 
print(is_palindrome(123))
```

```python
#3.default arguments
def sum_of_numbers(n1,n2,n3=0,n4=0):
    val = n1 + n2 + n3 + n4
    return val
#function call
print(sum_of_numbers(20,30))
print(sum_of_numbers(20,30,40))
print(sum_of_numbers(5,20,30,40))
```

```python
#4.variable/arbitrary length of positional arguments
def sum_of_nums(*args):
    #print(type(args))
    #print(args)
    s = 0
    for i in range(len(args)):
        s += args[i]
    return s
#function call
print(sum_of_nums())
print(sum_of_nums(10))
print(sum_of_nums(10,11,12))
print(sum_of_nums(10,11,12,13,14,15))
```

```python
#5. variable length of keyword arguments
def cost_of(**kwargs):
    total = 0
    for item in kwargs.items():
        #print(item)
        print(f"{item[0]} : {item[1]}")
        total += item[1]
    return total
#function call
print("Total:", cost_of(apple = 10, orange = 100, mango = 150))
```

```python
def L_pattern(n):
 for i in range(5):
    for j in range(6):
       if j==0  or i==n-1:
            print("*",end = " ")
       else:
           print(" ",end = " ")
    print()
L_pattern(5)
```

```python
def N_pattern(n):
 for i in range(5):
    for j in range(5):
       if j==0 or j==n-1 or i == j:
            print("*",end = " ")
       else:
            print(" ",end = " ")
    print()
N_pattern(5)
```

```python
def p_pattern(n):
 for i in range(5):
    for j in range(5):
       if j==0 or i==0 or i==n//2 or  (j==n-1 and i<n//2):
            print("*",end = " ")
       else:
            print(" ",end = " ")
    print()
p_pattern(5)
```
