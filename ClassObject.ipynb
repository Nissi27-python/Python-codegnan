```python
# car class definition
class Car():
    wheels = 4 #class attribute - same for each obj
    def __init__(self, color, price):
        # attributes
        self.color = color #instance attribute - diff for each obj
        self.price = price #instance attribute - diff for each obj
    # instance method definition
    def car_info(self):
        return f"The car is {self.color} and its price is {self.price}"
    # class method definition
    @classmethod
    def class_method(cls, new_wheels):
        # updating class attributes
        cls.wheels = new_wheels
        return f"{new_wheels} wheels updated for car class"
    #static method
    @staticmethod
    def static_method(speed):
        return f"This car moves with the speed of {speed} kmph"
# creating an object
honda = Car('Blue', 1000000)
tesla = Car('Red', 3500000)
print(honda.car_info())
print(honda.wheels)
print(honda.class_method(6))
print(honda.wheels, tesla.wheels)
print(honda.static_method(120))
```

```python
#finding the rectangle perimeter and area using oop
class Rectangle():
    def __init__(self,length,breadth):
        self.length = length
        self.breadth = breadth
    def area(self):
        return self.length * self.breadth
    def perimeter(self):
        return 2*(self.length + self.breadth)
rect = Rectangle(10,5)
print(rect.area())
print(rect.perimeter())

        
```
